```python
import pandas as pd

data_type = "validation"

df = pd.read_csv(f"/run/media/thomas/Elements/Preprocessed/{data_type}/metadata.csv")
```

```python
df.head()
```

```python
def update_file_path(row):
    # Check if the path contains "vessel"
    if "vessel" in row["path"]:
        # Replace the old directory with the new one
        new_path = row["path"].replace(
            "/onc/07_classified_wav_files/inclusion_4000_exclusion_6000/vessel",
            "/run/media/thomas/Elements/classified_files/inclusion_4000_exclusion_6000/vessel",
        )
    elif "background" in row["path"]:
        # If the path does not contain "vessel", it could be background. You can handle it accordingly.
        # For example, you might want to keep the same path or replace it with another directory.
        new_path = row["path"].replace(
            "/onc/07_classified_wav_files/inclusion_4000_exclusion_6000/background",
            "/run/media/thomas/Elements/classified_files/inclusion_4000_exclusion_6000/background",
        )

    return new_path
```

```python
df["path"] = df.apply(update_file_path, axis=1)
df = df[df["label"] != "other"]
df.head()
```

```python
df.to_csv(f"/run/media/thomas/Elements/Preprocessed/{data_type}/metadata_updated.csv")
```
